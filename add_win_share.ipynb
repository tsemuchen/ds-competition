```python
import pandas as pd

schools1 = ["michigan-state", "northwestern", "ohio-state"]
schools2 = ["maryland", "michigan", "minnesota"]
schools3 = ["indiana", "illinois", "nebraska"]
schools4 = ["iowa", "penn-state", "purdue"]
schools5 = ["rutgers", "wisconsin"]
years = list(range(2021, 2025))
```

```python
for s in schools5:
    winshare = pd.DataFrame()
    for y in years:
        df = pd.read_html(f"https://www.sports-reference.com/cbb/schools/{s}/men/{y}.html")
        data = df[11].iloc[:, [1, 2, 4] + list(range(20, 24))].copy()
        data["Year"] = y - 1
        winshare = pd.concat([winshare, data], ignore_index=True)
            
    winshare.to_csv(f"./ws/{s}_winshare.csv", index=False)
    print(f"{s} finished!")
```
